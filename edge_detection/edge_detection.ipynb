```python
#importing Libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading input image

img = cv2.imread("cubics.jpeg")
cv2.imshow('img_1', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#shape of image
img.shape
```

```python
#plotting image
plt.imshow(img)
```

```python
#converting colour image to gray_scale image

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#applying Canny Edge detection on gray scale image

img_canny = cv2.Canny(img_gray, 1, 320, apertureSize=5) #thesholds can be updated with help of matplot scale

cv2.imshow('img_2', img_canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#canny edge matrix

img_canny
```

```python
#applying Houghlines on canny image

lines = cv2.HoughLines(img_canny, 1, np.pi/180, 100)
```

```python
#output of houghlines operation has 2 parameters in the form of an array (rho value and theta value)

lines
```

```python
len(lines)
```

```python
# rho = x.cos(theta) + y.sin(theta)
# edges detection and plotting lines on each edges

if lines is not None:
    for i in range(len(lines)):
        rho = lines[i][0][0]
        theta = lines[i][0][1]
        
        a = np.cos(theta)
        b = np.sin(theta)
        
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))

        cv2.line(img,(x1,y1),(x2,y2),(155,155,255),1)
cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
